Notebook to ensure one box model code is working properly

```python
from methane_noise_forcing.noise import solve_ch4_onebox, solve_ch4_anomaly_lifetime
import matplotlib.pyplot as plt
import numpy as np
```

## test class with perturbation experiments
Tau and k and OH with lifetime of 10 years

```python
## test with tau directly
tau = 10 # years
C0 = 10 # ppb
t_eval = np.arange(0, 50, 0.5) # years

result = solve_ch4_onebox(
    t_eval=t_eval,
    C0=C0 * 1e-9,
    S=0,  # ppb/year
    tau=tau,
)

fig, ax = plt.subplots(figsize=(4, 3))

ax.plot(result.t, result.C * 1e9, label=f"τ={tau} yr")
ax.axhline(np.exp(-1) * C0, color='r', ls='-.', label=f"lifetime: {tau} yr")

ax.legend()
ax.set_xlabel("Time (years)")
ax.set_ylabel("Methane concentration (ppb)")
```

```python
## test with k and OH
C0 = 10 # ppb
t_eval = np.arange(0, 50, 0.5) # years
mean_lifetime = 10 # years
oh_mean = 1.3764384222381171e-13
k_ch4_oh = 1 / mean_lifetime / oh_mean

result = solve_ch4_onebox(
    t_eval=t_eval,
    C0=C0 * 1e-9,
    S=0,  # ppb/year
    k=k_ch4_oh, # (mol rxn / mol air)^-1 yr^-1
    OH=oh_mean, # (mol rxn / mol air)
)

fig, ax = plt.subplots(figsize=(4, 3))

ax.plot(result.t, result.C * 1e9, label=f"τ={tau} yr")
ax.axhline(np.exp(-1) * C0, color='r', ls='-.', label=f"lifetime: {tau} yr")

ax.legend()
ax.set_xlabel("Time (years)")
ax.set_ylabel("Methane concentration (ppb)")
```

```python
## finally, test with perturbation form
C0 = 10 # ppb
t_eval = np.arange(0, 50, 0.5) # years

result = solve_ch4_anomaly_lifetime(
    t_eval=t_eval,
    C0_anom=C0 * 1e-9,
    S_anom=0,  # ppb/year
    tau=10, # years
)

fig, ax = plt.subplots(figsize=(4, 3))

ax.plot(result.t, result.C * 1e9, label=f"τ={tau} yr")
ax.axhline(np.exp(-1) * C0, color='r', ls='-.', label=f"lifetime: {tau} yr")

ax.legend()
ax.set_xlabel("Time (years)")
ax.set_ylabel("Methane concentration (ppb)")
```
