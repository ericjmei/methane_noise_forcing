```python
## global parameters
variance_filtered_methane = 100  # ppb^2
taus_methane = np.arange(7, 13.2, 0.2)  # years

# parameters for sweeping forcing timescale
a, b = -2, 3  # log10 years
m = 10  # samples per decade
num = (b - a) * m + 1  # number of samples
taus_forcing = np.logspace(a, b, num)  # years

# lags to calculate
n_lag_years = 150  # number of year lags to calculate

# dts for calculating methane lags
dt_firn_orig = 0.5
dt_firn_narrow = 0.25
dt_firn_wide = 1.0
dts_firn = [dt_firn_narrow, dt_firn_orig, dt_firn_wide]
```

```python
## plotting functions
def plot_da_heat(
    da,
    *,
    ax=None,  # main plotting axes
    cbar_ax=None,  # optional separate axes for the colorbar
    cmap="Reds",
    log_color=False,  # True => LogNorm for both image & colorbar
    vmin=None,
    vmax=None,
    cbar_label=None,
    cbar_orientation="vertical",  # 'vertical' or 'horizontal'
    tight=False,
    **kwargs,
):
    """
    da: xarray.DataArray with dims ('taus_methane', 'taus_forcing')
    Returns: (h, ax, cbar)
    """
    if ax is None:
        ax = plt.gca()

    # Norm controls both the image mapping and the colorbar scale
    norm = mcolors.LogNorm(vmin=vmin, vmax=vmax) if log_color else None

    # Build kwargs for xarray's plot
    plot_kwargs = dict(
        x="taus_forcing",
        y="taus_methane",
        ax=ax,
        cmap=cmap,
        norm=norm,
        add_colorbar=True,
    )
    if cbar_ax is not None:
        plot_kwargs["cbar_ax"] = cbar_ax
    # Only pass cbar_kwargs if we actually set something, to avoid overriding defaults
    cbar_kwargs = {}
    if cbar_label is not None:
        cbar_kwargs["label"] = cbar_label
    if cbar_orientation != "vertical":
        cbar_kwargs["orientation"] = cbar_orientation
    if cbar_kwargs:
        plot_kwargs["cbar_kwargs"] = cbar_kwargs

    plot_kwargs.update(kwargs)
    h = da.plot(**plot_kwargs)

    # Axes formatting
    ax.set_xscale("log")
    ax.set_xlabel("Forcing timescale (yr)")
    ax.set_ylabel("Methane lifetime (yr)")
    ax.set_title("")

    # Decimal tick labels on a log x-axis (0.01, 0.1, 1, 10, ...)
    ax.xaxis.set_major_locator(mticker.LogLocator(base=10))
    ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda v, pos: f"{v:g}"))
    ax.xaxis.set_minor_formatter(mticker.NullFormatter())

    if tight:
        plt.tight_layout()

    # Grab the colorbar (xarray attaches it to the artist)
    cbar = h.colorbar
    return h, ax, cbar


def colorbar_decade_labels(cbar):
    vmin, vmax = cbar.norm.vmin, cbar.norm.vmax
    lo = int(np.ceil(np.log10(vmin)))
    hi = int(np.floor(np.log10(vmax)))
    ticks = [10**e for e in range(lo, hi + 1)]
    if ticks:  # only set if we actually have full decades in-range
        cbar.set_ticks(ticks)
    cbar.formatter = mticker.FuncFormatter(lambda v, pos: f"{v:g}")  # 1, 10, 100...
    cbar.update_ticks()
```

# generate and visualize kernel self-overlap (autocorrelation)

```python
firn_filter_orig = FirnFilter.fit_from_config(
    kernel_type=cfg.firn_kernel.type,
    parameters=cfg.firn_kernel.parameters,
    path_root=cfg.paths.firn_filters_root,
)
# firn_filter = FirnFilter.fit_gamma(
#     mode=30,
#     fwhm=27,
#     skew=0.7,
#     t_max=200,
#     dt=0.1,
#     taper_fraction=0
# )

dt = dt_firn_wide  # years
lags = np.arange(0, n_lag_years / dt, 1)  # years

# normalize kernel by ensuring unit area
# since series dt is equal to kernel dt, we can use the sum of the kernel directly
kernel_orig_normalized = firn_filter_orig.kernel / np.sum(firn_filter_orig.kernel)

# calculate self-overlap
c_taus = closed_form.calculate_kernel_self_lag(
    kernel=kernel_orig_normalized,
    lags=lags,
)

# plot
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# plot firn filter
time = np.arange(0, kernel_orig_normalized.shape[0] * dt, dt)
axs[0].plot(time, kernel_orig_normalized)
axs[0].set_xlabel("Time (years)")
axs[0].set_ylabel("Kernel (unitless)")

# plot self-overlap
axs[1].plot(lags * dt, c_taus)
axs[1].set_xlabel("Lag (years)")
axs[1].set_ylabel("Self-overlap (unitless)")

plt.tight_layout()
```

# two-timescale noise

## generate autocorrelations for range of timescales

```python
## iterate over methane lifetimes and forcing timescales to calculate autocorrelation over all lags
#  dimensions of output DataSet: (taus_methane, taus_forcing, lags)
autocorrelation_two_timescale_dict = {}  # bad, I know
for dt_firn in dts_firn:
    lags = np.arange(0, n_lag_years / dt_firn, 1)  # years
    autocorrelation_two_timescale = xr.DataArray(
        np.zeros((len(taus_methane), len(taus_forcing), len(lags))),
        dims=["taus_methane", "taus_forcing", "lags"],
        coords={
            "taus_methane": taus_methane,
            "taus_forcing": taus_forcing,
            "lags": lags,
        },
    )
    for tau_methane in taus_methane:
        for tau_forcing in taus_forcing:
            autocorrelation = closed_form.calculate_two_timescale_autocorrelation(
                phi_0=np.exp(-dt_firn / tau_methane),
                phi_1=np.exp(-dt_firn / tau_forcing),
                lags=lags,
            )
            autocorrelation_two_timescale.loc[
                dict(
                    taus_methane=tau_methane,
                    taus_forcing=tau_forcing,
                )
            ] = autocorrelation
    autocorrelation_two_timescale_dict[dt_firn] = autocorrelation_two_timescale
```

```python
## plot autocorrelations at specific year lags,
#  shouldn't change between different dt_firn since it's a function of lag in years, not lag in number of steps
lag_years = 3  # years
dt_firn = dt_firn_wide  # years
da_to_plot = autocorrelation_two_timescale_dict[dt_firn].sel(lags=(lag_years / dt_firn))

fig = plt.figure(figsize=(6, 5))
gs = fig.add_gridspec(ncols=2, nrows=1, width_ratios=[20, 1], wspace=0.05)
ax = fig.add_subplot(gs[0, 0])
cax = fig.add_subplot(gs[0, 1])

h, ax, cbar = plot_da_heat(
    da_to_plot,
    ax=ax,
    cbar_ax=cax,
)
h.set_clim(0.3, 1)
h.set_label("something")
ax.set_title(f"Autocorrelation at {lag_years} years lag")
cbar = h.colorbar
cbar.set_label("Autocorrelation")
```

## calculate variance required in atmospheric methane and forcing

```python
## iterate over methane lifetimes and forcing timescales to calculate atmospheric methane and forcing variances
#  dimensions of output DataSet: (taus_methane, taus_forcing)
atmospheric_methane_variance = xr.DataArray(
    np.zeros((len(taus_methane), len(taus_forcing), len(dts_firn))),
    dims=["taus_methane", "taus_forcing", "dts_firn"],
    coords={
        "taus_methane": taus_methane,
        "taus_forcing": taus_forcing,
        "dts_firn": dts_firn,
    },
)
forcing_variance = xr.DataArray(
    np.zeros((len(taus_methane), len(taus_forcing), len(dts_firn))),
    dims=["taus_methane", "taus_forcing", "dts_firn"],
    coords={
        "taus_methane": taus_methane,
        "taus_forcing": taus_forcing,
        "dts_firn": dts_firn,
    },
)
for dt_firn in dts_firn:
    for tau_methane in taus_methane:
        for tau_forcing in taus_forcing:
            # first, re-compute lags and self-overlap for this dt_firn (by changing dt, which re-indexes the kernel)
            lags = np.arange(0, n_lag_years / dt_firn, 1)  # years
            c_taus = closed_form.calculate_kernel_self_lag(
                kernel=kernel_orig_normalized,
                lags=lags,
            )

            variance_ratio_filtered_to_atmospheric = (
                closed_form.calculate_variance_ratio_filtered_to_instantaneous(
                    c_taus=c_taus,
                    instantaneous_autocorrelations=autocorrelation_two_timescale_dict[
                        dt_firn
                    ].loc[
                        dict(
                            taus_methane=tau_methane,
                            taus_forcing=tau_forcing,
                        )
                    ],
                )
            )
            variance_atmospheric_methane = (
                variance_filtered_methane / variance_ratio_filtered_to_atmospheric
            )
            atmospheric_methane_variance.loc[
                dict(
                    taus_methane=tau_methane,
                    taus_forcing=tau_forcing,
                    dts_firn=dt_firn,
                )
            ] = variance_atmospheric_methane

            variance_ratio_methane_to_forcing = closed_form.calculate_two_timescale_variance_ratio_methane_to_ar1_forcing(
                tau_methane=tau_methane,
                tau_forcing=tau_forcing,
            )
            variance_forcing = (
                variance_atmospheric_methane / variance_ratio_methane_to_forcing
            )
            forcing_variance.loc[
                dict(
                    taus_methane=tau_methane,
                    taus_forcing=tau_forcing,
                    dts_firn=dt_firn,
                )
            ] = variance_forcing

# convert units of variance from ppb^2/yr^2 to Tg^2/yr^2
# 1 ppb = 2.75 Tg of methane
conversion_factor = 2.75
variance_forcing *= conversion_factor**2
```

```python
## plot atmospheric methane variance required for a given filtered variance
dt_firn = dt_firn_narrow  # years

fig = plt.figure(figsize=(6, 5))
gs = fig.add_gridspec(ncols=2, nrows=1, width_ratios=[20, 1], wspace=0.05)
ax = fig.add_subplot(gs[0, 0])
cax = fig.add_subplot(gs[0, 1])

h, ax, cbar = plot_da_heat(
    atmospheric_methane_variance.sel(dts_firn=dt_firn),
    ax=ax,
    cbar_ax=cax,
    cmap="viridis",
)
cbar = h.colorbar
cbar.set_label("Atmospheric methane variance (ppb$^2$)")
h.set_clim(100, 230)
```

```python
## plot forcing variance required
dt_firn = dt_firn_narrow  # years

fig = plt.figure(figsize=(6, 5))
gs = fig.add_gridspec(ncols=2, nrows=1, width_ratios=[20, 1], wspace=0.05)
ax = fig.add_subplot(gs[0, 0])
cax = fig.add_subplot(gs[0, 1])

h, ax, cbar = plot_da_heat(
    forcing_variance.sel(dts_firn=dt_firn),
    ax=ax,
    cbar_ax=cax,
    cmap="viridis",
    log_color=True,
)
cbar = h.colorbar
cbar.set_label("Forcing variance (Tg$^2$ yr$^{-2}$)")
colorbar_decade_labels(cbar)
```

```python
## plotting functions
def annotate_panel(ax, text, loc="left", xpad=0.01, ypad=0.01, color="k", **kwargs):
    """Place bold text at the top-left/right inside an axis."""
    x = xpad if loc == "left" else 1 - xpad
    y = 1 - ypad
    ha = "left" if loc == "left" else "right"
    ax.text(
        x,
        y,
        text,
        transform=ax.transAxes,
        ha=ha,
        va="top",
        fontweight="bold",
        color=color,
        **kwargs,
    )


def compute_violin_coords(
    data,
    center,
    width,
    density_points=256,
    axis_scale="log",
    kde_kwargs=None,
    reflect_at_zero=False,
):
    """
    Compute violin plot coordinates for KDE density visualization.

    Parameters
    ----------
    data : array-like
        1D array of data values to visualize
    center : float
        Center position of the violin on the x-axis (e.g., timescale value)
    width : float
        Width parameter controlling violin spread:
        - For log scale: width in decades (log10 space)
        - For linear scale: absolute width in data units
    density_points : int, optional
        Number of points to sample for the violin shape
    axis_scale : {'log', 'linear'}, optional
        Type of x-axis scaling
    kde_kwargs : dict, optional
        Additional kwargs passed to gaussian_kde (e.g., bw_method)
    reflect_at_zero : bool, optional
        If True, reflect data at zero to prevent KDE from extending to negative values.
        Useful for distributions bounded at zero (like absolute values).

    Returns
    -------
    y : ndarray
        Y-coordinates (data values)
    x_left : ndarray
        Left x-coordinates for fill_betweenx
    x_right : ndarray
        Right x-coordinates for fill_betweenx
    """
    from scipy.stats import gaussian_kde

    data = np.asarray(data)
    data = data[~np.isnan(data)]

    if data.size == 0:
        return None, None, None

    # Handle reflection at zero for bounded distributions
    if reflect_at_zero:
        # Reflect negative values across zero
        data_for_kde = np.concatenate([data, -data])
    else:
        data_for_kde = data

    # Compute KDE
    kde_kwargs = kde_kwargs or {}
    kde = gaussian_kde(data_for_kde, **kde_kwargs)

    # Sample range
    dmin, dmax = np.min(data), np.max(data)
    y = np.linspace(dmin, dmax, density_points)

    # Evaluate KDE only at positive values
    v = kde(y)

    # Normalize to max = 1
    v = v / v.max()

    # Compute x coordinates based on scale
    if axis_scale == "log":
        # Width in decades (log10 space)
        half_width_decades = width * v
        m = 10.0**half_width_decades
        x_left = center / m
        x_right = center * m
    elif axis_scale == "linear":
        # Width in linear space
        half_width = width * v
        x_left = center - half_width
        x_right = center + half_width
    else:
        raise ValueError(f"axis_scale must be 'log' or 'linear', got {axis_scale}")

    return y, x_left, x_right


def multiplot_params(tau, tau_methane_mid, lw_other=0.8, alpha_other=0.7):
    """Ensures line with focus in panel b is thicker, more opaque, and on top"""
    if tau == tau_methane_mid:
        lw = 3
        alpha = 1
        zorder_val = 100
    else:
        lw = lw_other
        alpha = alpha_other
        zorder_val = 1 + 0.01 * tau
    kwargs = dict(lw=lw, alpha=alpha, zorder=zorder_val)
    return kwargs


## function for curve-following annotations
from matplotlib.textpath import TextPath
from matplotlib.patches import PathPatch
from matplotlib.font_manager import FontProperties
from matplotlib.path import Path


def text_along_curve(
    ax,
    x_data,
    y_data,
    text,
    x_start,
    fontsize=8,
    color="k",
    fontweight="normal",
    y_offset_points=0.0,  # <<--- NEW PARAMETER
    **patch_kwargs,
):
    """
    Draw `text` so that its baseline follows the curve (x_data, y_data),
    with an optional constant vertical offset expressed in points.
    """
    x_data = np.asarray(x_data)
    y_data = np.asarray(y_data)

    # --- 1. Transform curve to display coords ---
    xy_disp = ax.transData.transform(np.column_stack([x_data, y_data]))
    xd, yd = xy_disp[:, 0], xy_disp[:, 1]

    # Arc-length in display coords
    ds = np.hypot(np.diff(xd), np.diff(yd))
    s = np.concatenate([[0.0], np.cumsum(ds)])

    # Closest point in data space
    idx_start = np.argmin(np.abs(x_data - x_start))
    s_center = s[idx_start]

    # --- 2. Build text path in points ---
    fp = FontProperties(weight=fontweight)
    tp = TextPath((0, 0), text, size=fontsize, prop=fp)
    verts = tp.vertices.copy()
    codes = tp.codes

    # Points → pixels
    dpi = ax.figure.dpi
    pt_to_px = dpi / 72.0
    verts[:, 0] *= pt_to_px
    verts[:, 1] *= pt_to_px

    # Separate baseline (x-direction)
    x_min = verts[:, 0].min()
    verts[:, 0] -= x_min
    text_length = verts[:, 0].max() - verts[:, 0].min()

    # Center text baseline at s_center
    s_start = s_center - text_length / 2.0

    # --- NEW: vertical offset in display pixels ---
    y_offset_px = y_offset_points * pt_to_px

    # --- 3. Warp glyphs along curve ---
    verts_warped = np.zeros_like(verts)

    for i, (u, v) in enumerate(verts):
        s_target = s_start + u

        # Clamp
        if s_target <= 0:
            idx = 0
            t = 0.0
        elif s_target >= s[-1]:
            idx = len(s) - 2
            t = 1.0
        else:
            idx = np.searchsorted(s, s_target) - 1
            idx = np.clip(idx, 0, len(s) - 2)
            t = (s_target - s[idx]) / (s[idx + 1] - s[idx])

        # Interpolate curve point
        x_c = xd[idx] + t * (xd[idx + 1] - xd[idx])
        y_c = yd[idx] + t * (yd[idx + 1] - yd[idx])

        # Tangent
        tx = xd[idx + 1] - xd[idx]
        ty = yd[idx + 1] - yd[idx]
        nrm = np.hypot(tx, ty)
        if nrm == 0:
            tx, ty = 1.0, 0.0
            nx, ny = 0.0, 1.0
        else:
            tx /= nrm
            ty /= nrm
            nx, ny = -ty, tx  # normal vector

        # Warp vertex: baseline (u) maps to arclength, v maps normal
        x_v = x_c + v * nx
        y_v = y_c + v * ny

        # --- apply constant vertical offset in display coords ---
        y_v += y_offset_px

        verts_warped[i, 0] = x_v
        verts_warped[i, 1] = y_v

    # --- 4. Transform back to data coords ---
    inv = ax.transData.inverted()
    verts_data = inv.transform(verts_warped)

    # Build and add path patch
    text_path = Path(verts_data, codes)
    patch = PathPatch(
        text_path,
        facecolor=color,
        edgecolor="none",
        lw=0,
        **patch_kwargs,
    )
    ax.add_patch(patch)


import numpy as np


def arrow_along_curve(
    ax,
    x_data,
    y_data,
    x_start,
    length_points=40.0,
    offset_points=10.0,
    color="k",
    **arrowprops,
):
    """
    Draw an arrow roughly parallel to the curve (x_data, y_data),
    centered at x_start, with a constant offset in screen space.

    Parameters
    ----------
    ax : matplotlib.axes.Axes
    x_data, y_data : 1D arrays
        Curve in data coordinates.
    x_start : float
        Data x-location where the arrow is centered along the curve.
    length_points : float
        Arrow length in points (approx along the curve).
    offset_points : float
        Offset from the curve in points (normal to the curve).
    color : str
        Arrow color.
    arrowprops :
        Extra kwargs forwarded into ax.annotate(..., arrowprops=...).
    """
    x_data = np.asarray(x_data)
    y_data = np.asarray(y_data)

    # ---- Curve in display coords (handles log axes automatically) ----
    xy_disp = ax.transData.transform(np.column_stack([x_data, y_data]))
    xd, yd = xy_disp[:, 0], xy_disp[:, 1]

    # Arc length along the curve in display coords
    ds = np.hypot(np.diff(xd), np.diff(yd))
    s = np.concatenate([[0.0], np.cumsum(ds)])

    # Center position along curve: closest x to x_start in data space
    idx_center = np.argmin(np.abs(x_data - x_start))
    s_center = s[idx_center]

    # Convert from points to pixels
    dpi = ax.figure.dpi
    pt_to_px = dpi / 72.0
    L = length_points * pt_to_px
    d = offset_points * pt_to_px

    s_start = np.clip(s_center - L / 2.0, s[0], s[-1])
    s_end = np.clip(s_center + L / 2.0, s[0], s[-1])

    def point_and_normal_at_s(s_target):
        """Return (x_disp, y_disp) offset by d*normal, plus tangent."""
        if s_target <= s[0]:
            i = 0
            t = 0.0
        elif s_target >= s[-1]:
            i = len(s) - 2
            t = 1.0
        else:
            i = np.searchsorted(s, s_target) - 1
            i = np.clip(i, 0, len(s) - 2)
            t = (s_target - s[i]) / (s[i + 1] - s[i])

        # Curve point
        x_c = xd[i] + t * (xd[i + 1] - xd[i])
        y_c = yd[i] + t * (yd[i + 1] - yd[i])

        # Tangent
        tx = xd[i + 1] - xd[i]
        ty = yd[i + 1] - yd[i]
        nrm = np.hypot(tx, ty)
        if nrm == 0:
            tx, ty = 1.0, 0.0
        else:
            tx /= nrm
            ty /= nrm

        # Normal (rotate tangent 90° CCW)
        nx, ny = -ty, tx

        # Offset point in display space
        x_o = x_c + d * nx
        y_o = y_c + d * ny

        return x_o, y_o, tx, ty

    # Start and end of the arrow in display coords
    x_start_disp, y_start_disp, tx0, ty0 = point_and_normal_at_s(s_start)
    x_end_disp, y_end_disp, tx1, ty1 = point_and_normal_at_s(s_end)

    # (Optional) make the arrow point roughly "up-left" by swapping if needed
    # You can flip direction if you want:
    # if x_end_disp > x_start_disp:
    #     x_start_disp, y_start_disp, x_end_disp, y_end_disp = \
    #         x_end_disp, y_end_disp, x_start_disp, y_start_disp

    # Back to data coords
    inv = ax.transData.inverted()
    x0, y0 = inv.transform([x_start_disp, y_start_disp])
    x1, y1 = inv.transform([x_end_disp, y_end_disp])

    # Draw arrow
    props = dict(arrowstyle="<-", color=color, linewidth=0.8)
    props.update(arrowprops)

    ax.annotate("", xy=(x1, y1), xytext=(x0, y0), arrowprops=props)
```

```python
import numpy as np
from matplotlib.colors import to_rgba

LN10 = np.log(10.0)


def lognorm_sigma_log10_from_mean_std(mean, std):
    """
    Convert (mean, std) in linear space to sigma in log10-space
    for a lognormal distribution.
    """
    var = std**2
    sigma_ln = np.sqrt(np.log(var / mean**2 + 1.0))
    sigma_log10 = sigma_ln / LN10
    return sigma_log10


import numpy as np
from matplotlib.colors import to_rgba, LinearSegmentedColormap

LN10 = np.log(10.0)


def add_loglog_gaussian_blob(
    ax,
    *,
    x_mean,
    x_sigma_log10,
    y_mean,
    y_sigma_log10,
    color,
    max_alpha=0.65,
    n_grid=220,
    n_sigma=6.0,
    sigma_mult=2.2,
    taper_gamma=0.5,
    alpha_floor=1e-3,  # <- key: mask tiny tails so they don't tint the whole plot
    zorder=0.8,
):
    if x_mean <= 0 or y_mean <= 0:
        raise ValueError("Means must be > 0 for log axes.")
    if x_sigma_log10 <= 0 or y_sigma_log10 <= 0:
        raise ValueError("Sigmas must be > 0.")

    mx = np.log10(x_mean)
    my = np.log10(y_mean)

    sx = x_sigma_log10 * sigma_mult
    sy = y_sigma_log10 * sigma_mult

    x_log = np.linspace(mx - n_sigma * sx, mx + n_sigma * sx, n_grid)
    y_log = np.linspace(my - n_sigma * sy, my + n_sigma * sy, n_grid)
    XLOG, YLOG = np.meshgrid(x_log, y_log)

    # 2D Gaussian in log10 space
    dx2 = ((XLOG - mx) / sx) ** 2
    dy2 = ((YLOG - my) / sy) ** 2
    A = np.exp(-0.5 * (dx2 + dy2))

    # gentler taper (fatter blob)
    A = (A**taper_gamma) * max_alpha

    # mask tiny alpha so mesh doesn't cover the whole plot
    A = np.ma.masked_less(A, alpha_floor)

    # to linear coords
    X = 10**XLOG
    Y = 10**YLOG

    # colormap: transparent -> colored (with alpha baked in)
    r, g, b, _ = to_rgba(color)
    cmap = LinearSegmentedColormap.from_list(
        "blob",
        [(r, g, b, 0.0), (r, g, b, max_alpha)],
        N=256,
    )

    levels = np.linspace(alpha_floor, max_alpha, 100)

    ax.contourf(
        X,
        Y,
        A,
        levels=levels,
        cmap=cmap,
        antialiased=True,
        zorder=zorder,
    )
```
