```python
from methane_noise_forcing import generate_ar1_noise, generate_ar1_noise_using_tau
import numpy as np
import matplotlib.pyplot as plt
```

```python
# test autocorrelation function
tau_days = 1e2 # 0 tau days means white noise
dt_days = 1
variance = 15 # arb. unit
n_tau_steps = 10000

# generate AR(1) noise
noise = generate_ar1_noise_using_tau(tau_days, dt_days, variance, n_tau_steps, seed=42)

# calculate 0-lag variance
var_0_lag = np.var(noise, ddof=1)

# plot
plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(noise)), noise, label='AR(1) noise', color='blue')
plt.title('AR(1) Noise with Tau = {}'.format(tau_days))
plt.xlabel('Time Steps')
plt.ylabel('Noise Value')
plt.axhline(0, color='black', lw=0.5, ls='--')
# write prescribed and actual variance on plot
plt.text(0.5, 16, f'Prescribed Variance: {variance}', fontsize=12, color='k')
plt.text(0.5, 14, f'Actual Variance: {var_0_lag:.2f}', fontsize=12, color='k')
```
